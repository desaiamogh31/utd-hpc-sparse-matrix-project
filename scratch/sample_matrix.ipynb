```python
import ssgetpy
import tarfile
from pathlib import Path
from scipy.io import mmread, mmwrite

# 1. Search for one small matrix
result = ssgetpy.search(nzbounds=(1000, 10000), isspd=False, limit=1)
small_matrix = result[0]
print(small_matrix)

# 2. Download archive
files = small_matrix.download(destpath="matrices")
archive = Path(files[0])
print("Downloaded:", archive)

# 3. Extract archive
extract_dir = archive.parent / archive.stem
extract_dir.mkdir(exist_ok=True)

with tarfile.open(archive, "r:gz") as tar:
    tar.extractall(extract_dir)

# 4. Find the Matrix Market file
mtx_file = next(extract_dir.rglob("*.mtx"))
print("Matrix file:", mtx_file)

# 5. Load into SciPy sparse matrix
A = mmread(str(mtx_file)).tocsr()
print("Shape:", A.shape)
print("NNZ:", A.nnz)

# 6. Store a local copy wherever you want
output_path = Path("matrices/small_matrix.mtx")
mmwrite(str(output_path), A)
print("Saved to:", output_path)
```
